```python
def passByReference(list):
  list.extend([40, 50, 60])
  print ("List inside the function : {}".format(list))
  return

list=[10, 20, 30]
passByReference(list)
print ("List outside the function : ", list)
```

```python
def passByValue(list):
  list = [40, 50, 60]
  print ("List inside the function : {}".format(list))
  return

list=[10, 20, 30]
passByValue(list)
print ("List outside the function : ", list)
```
